```python
import requests
import time
import random
import os
```

```python
def get_img(num,year):
    url="http://202.120.126.33/pyxx/Imagefor.aspx?xh=%s"%(num)
    cookies={"ASP.NET_SessionId":"shzexxlb4caoankmcstatxja"}
    res = requests.get(url,cookies=cookies)
    print("get %s"%(num))
    if len(res.text)>600:
        with open("/home/xutao/Pictures/shuImg/%s/%s.jpg"%(year,num),'wb') as f:
            f.write(res.content)
        return True
    return False
```

```python
# get_img("18721771")
year=['18','17']
unk=['72','12']
for j in year:
    path="/home/xutao/Pictures/shuImg/%s"%(j)
    if not os.path.exists(path):
        os.makedirs(path) 
    for i in range(1,9999):
        for k in unk:
            s=j+k+'%04d'%i
            if not get_img(s,j):
                break
            time.sleep(random.randint(1, 2))
```
